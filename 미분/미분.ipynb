- lim h->0 은 극한을 나타내며, h 가 한없이 0에 근접한다는 뜻이다. h 를 한없이 0에 가깝게 줄여 x 의 변화 비율을 구하면 그 값이 바로 y = f(x) 의 미분이다. 또한 y = f(x) 가 어떤 구간에서 미분 가능하다면, 해당구간의 '모든 x' 에서 성립한다. 따라서 위의 식의 f`(x) 도 함수이며, f(x) 의 도함수라고 한다.
